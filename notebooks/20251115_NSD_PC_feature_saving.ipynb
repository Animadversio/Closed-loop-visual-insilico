```python
import torch
import pickle as pkl
from os.path import join

# Define paths
outputroot = r"/n/holylabs/LABS/alvarez_lab/Lab/VVS_Accentuation/Encoding_models"
subject_id = "three0_250426-250501"  # or "leap_250426-250501", etc.
modelname = "resnet50_robust"  # or any other model
output_subdir_name = "model_outputs_pca4all"
figdir = join(outputroot, subject_id, output_subdir_name)
# 1. Load the PCA/feature transformers (Xtfmer_lyrswp)
# Try pickle first (works for most models)
try:
    Xtfmer_lyrswp = pkl.load(open(join(figdir, 
        f"{subject_id}_{modelname}_sweep_regressors_layers_Xtfmer_RidgeCV.pkl"), "rb"))
    print(f"Loaded Xtfmer from pickle")
except:
    # For ResNet50 variants, try torch file (sparse version)
    Xtfmer_lyrswp = torch.load(join(figdir, 
        f"{subject_id}_{modelname}_sweep_regressors_layers_Xtfmer_RidgeCV.pth"))
    print(f"Loaded Xtfmer from torch file")

# 2. Load the fitted regression models
fit_models_lyrswp = torch.load(join(figdir, 
    f"{subject_id}_{modelname}_sweep_regressors_layers_fitmodels_RidgeCV.pth"))

# 3. Load prediction metadata (D2 scores per unit)
pred_D2_dict = pkl.load(open(join(figdir, 
    f"{subject_id}_{modelname}_sweep_regressors_layers_pred_meta.pkl"), "rb"))

print(f"\nXtfmer keys (sample): {list(Xtfmer_lyrswp.keys())[:3]}")
print(f"Fit models keys (sample): {list(fit_models_lyrswp.keys())[:3]}")
print(f"Pred dict keys: {list(pred_D2_dict.keys())}")
```
